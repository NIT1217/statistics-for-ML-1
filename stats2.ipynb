1. What is a random variable in probability theory?
A random variable in probability theory is a variable whose possible values are numerical outcomes of a random phenomenon.

2. What are the types of random variables?
The types of random variables are discrete random variables and continuous random variables.

3. What is the difference between discrete and continuous distributions?
Discrete distributions deal with random variables that can only take on a finite or countably infinite number of distinct values, often integers (e.g., number of heads in coin flips). Continuous distributions, on the other hand, deal with random variables that can take on any value within a given range (e.g., height or weight).

4. What are probability distribution functions (PDF)?
Probability Distribution Functions (PDFs) are functions that describe the likelihood of a continuous random variable taking on a given value. For continuous variables, the PDF does not give the probability of a specific value but rather the probability density, where the area under the curve between two points represents the probability of the variable falling within that range. For discrete variables, the analogous concept is the Probability Mass Function (PMF).

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
Cumulative Distribution Functions (CDFs) differ from Probability Distribution Functions (PDFs) in that a CDF gives the probability that a random variable will take a value less than or equal to a given value, while a PDF describes the probability density for continuous variables or the probability for discrete variables at a specific value.

6. What is a discrete uniform distribution?
A discrete uniform distribution is a probability distribution where a finite number of values are equally likely to be observed. Each value has the same probability.

7. What are the key properties of a Bernoulli distribution?
The key properties of a Bernoulli distribution are that it models a single trial with only two possible outcomes: success (usually denoted as 1) or failure (usually denoted as 0). The probability of success is denoted by p, and the probability of failure is 1−p.

8. What is the binomial distribution, and how is it used in probability?
The binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials, where each trial has only two possible outcomes (success or failure) and the probability of success remains constant for each trial. It is used in probability to model scenarios where we are interested in the number of times an event occurs in a series of identical experiments.

9. What is the Poisson distribution and where is it applied?
The Poisson distribution is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space if these events occur with a known constant mean rate and independently of the time since the last event. It is applied in various fields to model rare events, such as the number of phone calls received by a call center in an hour, the number of defects per unit area in manufacturing, or the number of car accidents at an intersection in a day.

10. What is a continuous uniform distribution?
A continuous uniform distribution is a probability distribution where all values within a given interval are equally likely. The probability density function is constant over the interval and zero elsewhere.

11. What are the characteristics of a normal distribution?
The characteristics of a normal distribution include its bell-shaped, symmetric curve. The mean, median, and mode are all equal and located at the center of the distribution. The spread of the distribution is determined by its standard deviation, and approximately 68% of the data falls within one standard deviation of the mean, 95% within two standard deviations, and 99.7% within three standard deviations.

12. What is the standard normal distribution, and why is it important?
The standard normal distribution is a special case of the normal distribution with a mean of 0 and a standard deviation of 1. It is important because it allows for the standardization of any normal distribution, enabling comparisons between different datasets and the use of standard normal tables (Z-tables) to find probabilities.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
The Central Limit Theorem (CLT) states that, given a sufficiently large sample size from any population with a finite mean and variance, the sampling distribution of the sample mean will be approximately normally distributed, regardless of the original distribution's shape. It is critical in statistics because it allows us to apply normal distribution theory to sample means, even when the population distribution is not normal, which is fundamental for hypothesis testing and constructing confidence intervals.

14. How does the Central Limit Theorem relate to the normal distribution?
The Central Limit Theorem (CLT) directly relates to the normal distribution by stating that as the sample size increases, the sampling distribution of the sample mean will approach a normal distribution, regardless of the shape of the original population distribution[cite: 10]. This means that even if the underlying data is not normally distributed, the distribution of sample means from that data will tend towards normality if the samples are large enough.

15. What is the application of Z statistics in hypothesis testing?
Z-statistics are applied in hypothesis testing to determine how many standard deviations an observed sample mean is from the hypothesized population mean. They are used when the population standard deviation is known or when the sample size is large enough (typically n > 30) for the Central Limit Theorem to apply, allowing for the use of the standard normal distribution to calculate p-values and make decisions about the null hypothesis.

16. How do you calculate a Z-score, and what does it represent?
A Z-score is calculated using the formula: Z=(x−μ)/σ, where x is the individual data point, μ is the population mean, and σ is the population standard deviation. A Z-score represents the number of standard deviations an individual data point is away from the mean of the distribution. A positive Z-score indicates the data point is above the mean, while a negative Z-score indicates it is below the mean.

17. What are point estimates and interval estimates in statistics?
Point estimates in statistics are single values used to approximate a population parameter (e.g., sample mean as an estimate for population mean). Interval estimates, on the other hand, provide a range of values within which the population parameter is likely to lie, along with a certain level of confidence (e.g., confidence interval).

18. What is the significance of confidence intervals in statistical analysis?
Confidence intervals are significant in statistical analysis because they provide a range of plausible values for an unknown population parameter, rather than just a single point estimate. They also quantify the uncertainty associated with the estimate, indicating the precision of the estimation and allowing researchers to make more informed decisions and interpretations about the population based on sample data.

19. What is the relationship between a Z-score and a confidence interval?
Z-scores are used in the calculation of confidence intervals, particularly when the population standard deviation is known or the sample size is large. The critical Z-values (e.g., Z = 1.96 for a 95% confidence interval) define the boundaries of the confidence interval around the sample mean, indicating how many standard errors away from the mean the interval extends to achieve a desired level of confidence.

20. How are Z-scores used to compare different distributions?
Z-scores are used to compare different distributions by standardizing the data. By converting raw scores from different distributions into Z-scores, one can compare individual data points on a common scale, regardless of their original means and standard deviations. This allows for meaningful comparisons of relative position within their respective distributions.

21. What are the assumptions for applying the Central Limit Theorem?
The assumptions for applying the Central Limit Theorem are:

The samples must be independent and identically distributed (i.i.d.).
The sample size must be sufficiently large (generally n≥30 is a common guideline, though it can be smaller for distributions that are already somewhat symmetric or larger for highly skewed distributions).
The population must have a finite mean and a finite variance.
22. What is the concept of expected value in a probability distribution?
The expected value (or expectation) in a probability distribution is the weighted average of all possible values a random variable can take, where the weights are the probabilities of each value occurring. It represents the long-run average outcome if the experiment were repeated many times.

23. How does a probability distribution relate to the expected outcome of a random variable?
A probability distribution relates to the expected outcome of a random variable by providing the probabilities associated with each possible outcome. The expected outcome (expected value) is then calculated as the sum of each possible outcome multiplied by its corresponding probability, effectively representing the long-term average value of the random variable if the process were repeated infinitely.

```python
# 2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

import numpy as np
import matplotlib.pyplot as plt

# Define the range for the discrete uniform distribution
low = 1
high = 6  # For example, outcomes of a fair die

# Generate possible outcomes
outcomes = np.arange(low, high + 1)

# Calculate the probability for each outcome
# For a discrete uniform distribution, each outcome has an equal probability
probability = 1 / len(outcomes)
pmf = np.full(len(outcomes), probability)

# Plot the PMF
plt.figure(figsize=(8, 6))
plt.bar(outcomes, pmf, tick_label=outcomes, color='skyblue')
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.title("Discrete Uniform Distribution PMF")
plt.xticks(outcomes)
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# 3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(k, p):
    """
    Calculates the Probability Mass Function (PMF) of a Bernoulli distribution.
    For Bernoulli, it's typically called PMF, not PDF.

    Args:
        k (int): The outcome (0 for failure, 1 for success).
        p (float): The probability of success (0 <= p <= 1).

    Returns:
        float: The probability of outcome k.
    """
    if not (0 <= p <= 1):
        raise ValueError("Probability 'p' must be between 0 and 1.")
    if k == 1:
        return p
    elif k == 0:
        return 1 - p
    else:
        return 0  # Outcomes other than 0 or 1 have zero probability

# Example usage:
p_success = 0.7
prob_success = bernoulli_pdf(1, p_success)
prob_failure = bernoulli_pdf(0, p_success)

print(f"Probability of success (k=1) for p={p_success}: {prob_success}")
print(f"Probability of failure (k=0) for p={p_success}: {prob_failure}")
print(f"Probability of k=2 for p={p_success}: {bernoulli_pdf(2, p_success)}")
```

```python
# 6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.

import numpy as np
import matplotlib.pyplot as plt

# Define the range for the discrete uniform distribution
low = 1
high = 6  # Example: outcomes of a fair die

# Possible outcomes
outcomes = np.arange(low, high + 1)

# Calculate the probability for each outcome (PMF)
probability_per_outcome = 1 / len(outcomes)

# Calculate the CDF
cdf_values = []
cumulative_prob = 0
for i in range(len(outcomes)):
    cumulative_prob += probability_per_outcome
    cdf_values.append(cumulative_prob)

# Plot the CDF
plt.figure(figsize=(8, 6))
plt.step(outcomes, cdf_values, where='post', color='darkblue', marker='o', linestyle='-')
plt.xlabel("Outcome (x)")
plt.ylabel("P(X <= x)")
plt.title("Cumulative Distribution Function (CDF) of Discrete Uniform Distribution")
plt.xticks(outcomes)
plt.yticks(np.linspace(0, 1, 11))
plt.grid(True, linestyle='--')
plt.show()
```

```python
# 9. Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_and_plot_z_scores(data):
    """
    Calculates Z-scores for a given dataset and plots their distribution against a standard normal distribution.

    Args:
        data (array-like): The input dataset.
    """
    data = np.array(data) # Ensure it's a NumPy array
    mean = np.mean(data)
    std_dev = np.std(data)

    if std_dev == 0:
        print("Standard deviation is zero. Cannot calculate Z-scores.")
        return

    z_scores = (data - mean) / std_dev

    # Plotting
    plt.figure(figsize=(12, 6))

    # Histogram of Z-scores
    plt.subplot(1, 2, 1)
    plt.hist(z_scores, bins=30, density=True, alpha=0.7, color='orange', edgecolor='black')
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.title("Histogram of Calculated Z-scores")
    plt.grid(axis='y', linestyle='--')

    # Plot Z-scores against standard normal PDF
    plt.subplot(1, 2, 2)
    x = np.linspace(-4, 4, 100) # Range for standard normal
    plt.plot(x, norm.pdf(x), color='red', linestyle='--', label='Standard Normal PDF')
    plt.hist(z_scores, bins=30, density=True, alpha=0.6, color='orange', edgecolor='black', label='Z-score Histogram')
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.title("Z-scores vs. Standard Normal Distribution")
    plt.legend()
    plt.grid(axis='y', linestyle='--')

    plt.tight_layout()
    plt.show()

    print(f"Original Data Mean: {mean:.2f}")
    print(f"Original Data Standard Deviation: {std_dev:.2f}")
    print(f"Mean of Z-scores: {np.mean(z_scores):.2f}")
    print(f"Standard Deviation of Z-scores: {np.std(z_scores):.2f}")

# Example usage with some arbitrary data
data_sample = np.random.normal(loc=100, scale=15, size=500) # Normally distributed data
# data_sample = np.random.exponential(scale=2, size=500) # Non-normal data to see effect
calculate_and_plot_z_scores(data_sample)
```

```python
# 10. Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

import numpy as np
import matplotlib.pyplot as plt

# Define a non-normal distribution (e.g., Exponential distribution)
# The population mean (lambda in exponential)
population_lambda = 2
population_mean = population_lambda
population_std_dev = population_lambda

# Parameters for the CLT simulation
sample_size = 30 # n >= 30 is generally considered "large enough" for CLT
num_samples = 10000 # Number of times we draw samples and calculate their means

# Generate means of samples
sample_means = []
for _ in range(num_samples):
    # Draw a sample of size 'sample_size' from the exponential distribution
    sample = np.random.exponential(scale=population_lambda, size=sample_size)
    # Calculate the mean of this sample
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)

# Convert to NumPy array for easier calculations
sample_means = np.array(sample_means)

# Plotting
plt.figure(figsize=(14, 6))

# Plot the original (non-normal) population distribution
plt.subplot(1, 2, 1)
population_data = np.random.exponential(scale=population_lambda, size=100000) # Large population for visualization
plt.hist(population_data, bins=50, density=True, alpha=0.7, color='skyblue', edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Density")
plt.title(f"Original Population Distribution (Exponential, λ={population_lambda})")
plt.grid(axis='y', linestyle='--')

# Plot the distribution of sample means
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=50, density=True, alpha=0.7, color='lightgreen', edgecolor='black')
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.title(f"Distribution of Sample Means (n={sample_size}, {num_samples} samples)")
plt.grid(axis='y', linestyle='--')

# Add theoretical mean and standard deviation of sample means
expected_mean_of_means = population_mean
expected_std_dev_of_means = population_std_dev / np.sqrt(sample_size)

plt.axvline(expected_mean_of_means, color='red', linestyle='dashed', linewidth=2, label=f'Theoretical Mean of Means: {expected_mean_of_means:.2f}')
plt.text(expected_mean_of_means + 0.1, plt.ylim()[1] * 0.9, f'Mean: {np.mean(sample_means):.2f}', color='red')
plt.text(expected_mean_of_means + 0.1, plt.ylim()[1] * 0.8, f'Std Dev: {np.std(sample_means):.2f}', color='red')

plt.legend()
plt.tight_layout()
plt.show()

print(f"Population Mean (Exponential): {population_mean:.2f}")
print(f"Population Standard Deviation (Exponential): {population_std_dev:.2f}")
print(f"Mean of Sample Means: {np.mean(sample_means):.2f}")
print(f"Standard Deviation of Sample Means (Standard Error): {np.std(sample_means):.2f}")
print(f"Theoretical Standard Error (Population Std Dev / sqrt(n)): {expected_std_dev_of_means:.2f}")
```

```python
# 16. Write a Python function to calculate and plot the standard normal distribution (mean=0, std=1).

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_standard_normal_distribution():
    """
    Calculates and plots the Probability Density Function (PDF)
    of the standard normal distribution (mean=0, std=1).
    """
    mean = 0
    std_dev = 1

    # Generate x values (Z-scores)
    x = np.linspace(-4, 4, 1000) # A common range for Z-scores

    # Calculate the PDF values for the standard normal distribution
    pdf_values = norm.pdf(x, loc=mean, scale=std_dev)

    # Plotting
    plt.figure(figsize=(8, 6))
    plt.plot(x, pdf_values, color='darkgreen', linewidth=2)
    plt.fill_between(x, 0, pdf_values, color='lightgreen', alpha=0.5) # Shade the area under the curve
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.title(f"Standard Normal Distribution (Mean={mean}, Std Dev={std_dev})")
    plt.grid(True, linestyle='--')
    plt.axvline(0, color='gray', linestyle=':', label='Mean = 0')
    plt.legend()
    plt.show()

# Call the function to plot the standard normal distribution
plot_standard_normal_distribution()
```

```python
# 18. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_and_compare_z_score(data_point, dataset):
    """
    Calculates the Z-score for a given data point within a dataset
    and visualizes its position relative to the standard normal distribution.

    Args:
        data_point (float): The specific data point for which to calculate the Z-score.
        dataset (array-like): The dataset from which the mean and standard deviation are calculated.
    """
    dataset = np.array(dataset)
    mean = np.mean(dataset)
    std_dev = np.std(dataset)

    if std_dev == 0:
        print("Standard deviation of the dataset is zero. Cannot calculate Z-score.")
        return

    z_score = (data_point - mean) / std_dev

    print(f"Dataset Mean: {mean:.2f}")
    print(f"Dataset Standard Deviation: {std_dev:.2f}")
    print(f"Data Point: {data_point:.2f}")
    print(f"Calculated Z-score: {z_score:.4f}")

    # Plotting the standard normal distribution
    plt.figure(figsize=(10, 6))
    x = np.linspace(-4, 4, 1000)
    pdf_values = norm.pdf(x, loc=0, scale=1) # Standard normal PDF

    plt.plot(x, pdf_values, color='blue', linewidth=2, label='Standard Normal PDF')
    plt.fill_between(x, 0, pdf_values, color='skyblue', alpha=0.3)

    # Mark the calculated Z-score on the plot
    plt.axvline(z_score, color='red', linestyle='--', linewidth=2, label=f'Your Z-score: {z_score:.2f}')
    plt.scatter(z_score, norm.pdf(z_score), color='red', s=100, zorder=5) # Mark the point

    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.title("Z-score Comparison to Standard Normal Distribution")
    plt.grid(True, linestyle='--')
    plt.axvline(0, color='gray', linestyle=':', label='Mean = 0')
    plt.legend()
    plt.show()

# Example usage:
# Create a sample dataset (e.g., student scores)
student_scores = [75, 82, 90, 68, 70, 85, 95, 78, 88, 72, 65, 92, 80, 83, 79]
my_score = 90

calculate_and_compare_z_score(my_score, student_scores)

# Another example with different data
heights_cm = [170, 175, 165, 180, 160, 185, 172, 168, 178, 173]
person_height = 160

calculate_and_compare_z_score(person_height, heights_cm)
```

```python
# 19. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy import stats

def z_test_for_mean(sample_mean, population_mean, population_std_dev, sample_size, alpha=0.05, alternative='two-sided'):
    """
    Performs a Z-test for a population mean.

    Args:
        sample_mean (float): The mean of the sample.
        population_mean (float): The hypothesized population mean (null hypothesis).
        population_std_dev (float): The known population standard deviation.
        sample_size (int): The number of observations in the sample.
        alpha (float, optional): The significance level. Defaults to 0.05.
        alternative (str, optional): The alternative hypothesis.
            'two-sided': H1: mu != mu0
            'less': H1: mu < mu0
            'greater': H1: mu > mu0
            Defaults to 'two-sided'.

    Returns:
        tuple: A tuple containing the Z-statistic, p-value, and a decision on the null hypothesis.
    """
    if sample_size <= 0:
        raise ValueError("Sample size must be greater than 0.")
    if population_std_dev <= 0:
        raise ValueError("Population standard deviation must be positive.")

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

    # Calculate the p-value based on the alternative hypothesis
    if alternative == 'two-sided':
        p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
    elif alternative == 'less':
        p_value = stats.norm.cdf(z_statistic)
    elif alternative == 'greater':
        p_value = 1 - stats.norm.cdf(z_statistic)
    else:
        raise ValueError("Invalid 'alternative' specified. Choose 'two-sided', 'less', or 'greater'.")

    # Make a decision
    if p_value < alpha:
        decision = "Reject the null hypothesis (H0)"
    else:
        decision = "Fail to reject the null hypothesis (H0)"

    return z_statistic, p_value, decision

# --- Example Usage ---

# Scenario: A manufacturer claims that the average lifespan of their light bulbs is 10,000 hours with a standard deviation of 500 hours.
# A consumer group tests 100 bulbs and finds their average lifespan to be 9,850 hours.
# Is there enough evidence at a 5% significance level to say the average lifespan is less than 10,000 hours?

# Given:
sample_mean_observed = 9850  # Sample mean lifespan
population_mean_hypothesized = 10000 # Hypothesized population mean (H0)
population_std_dev_known = 500 # Known population standard deviation
n_sample_size = 100 # Sample size
alpha_level = 0.05 # Significance level

print("--- Z-Test for Light Bulb Lifespan (One-tailed: less) ---")
z_stat, p_val, decision_result = z_test_for_mean(
    sample_mean=sample_mean_observed,
    population_mean=population_mean_hypothesized,
    population_std_dev=population_std_dev_known,
    sample_size=n_sample_size,
    alpha=alpha_level,
    alternative='less'
)

print(f"Calculated Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(f"Significance Level (alpha): {alpha_level}")
print(f"Decision: {decision_result}")
if decision_result.startswith("Reject"):
    print("Conclusion: There is sufficient evidence to suggest the average lifespan is less than 10,000 hours.")
else:
    print("Conclusion: There is not sufficient evidence to suggest the average lifespan is less than 10,000 hours.")


print("\n--- Z-Test for Light Bulb Lifespan (Two-tailed example) ---")
# If the question was: Is there enough evidence to say the average lifespan is *different* from 10,000 hours?
z_stat_two_sided, p_val_two_sided, decision_two_sided = z_test_for_mean(
    sample_mean=sample_mean_observed,
    population_mean=population_mean_hypothesized,
    population_std_dev=population_std_dev_known,
    sample_size=n_sample_size,
    alpha=alpha_level,
    alternative='two-sided'
)
print(f"Calculated Z-statistic: {z_stat_two_sided:.4f}")
print(f"P-value: {p_val_two_sided:.4f}")
print(f"Significance Level (alpha): {alpha_level}")
print(f"Decision: {decision_two_sided}")
```

```python
# 20. Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
from scipy import stats

def calculate_confidence_interval_mean(data, confidence_level=0.95):
    """
    Calculates the confidence interval for the mean of a dataset.
    Assumes population standard deviation is unknown, so uses t-distribution.
    If sample size is large (n>30), t-distribution approximates Z-distribution.

    Args:
        data (array-like): The input dataset.
        confidence_level (float, optional): The desired confidence level (e.g., 0.95 for 95%).
                                            Defaults to 0.95.

    Returns:
        tuple: A tuple containing the sample mean, margin of error, and the confidence interval (lower_bound, upper_bound).
    """
    data = np.array(data)
    n = len(data)
    if n <= 1:
        raise ValueError("Dataset must contain more than one element to calculate a confidence interval.")

    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1) # Use ddof=1 for sample standard deviation

    # Calculate the t-critical value (for unknown population std dev)
    # Degrees of freedom: n - 1
    t_critical = stats.t.ppf((1 + confidence_level) / 2, df=n - 1)

    # Calculate the standard error of the mean
    standard_error = sample_std / np.sqrt(n)

    # Calculate the margin of error
    margin_of_error = t_critical * standard_error

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return sample_mean, margin_of_error, (lower_bound, upper_bound)

# --- Example Usage ---

# Sample dataset (e.g., heights of students in cm)
heights = [165, 170, 168, 175, 172, 180, 169, 171, 178, 173, 167, 174, 166, 176, 170, 179, 172, 175, 169, 171]

# Calculate a 95% confidence interval
conf_level = 0.95
mean, moe, ci = calculate_confidence_interval_mean(heights, confidence_level=conf_level)

print(f"Sample Mean: {mean:.2f} cm")
print(f"Confidence Level: {conf_level * 100}%")
print(f"Margin of Error: {moe:.2f} cm")
print(f"Confidence Interval: ({ci[0]:.2f} cm, {ci[1]:.2f} cm)")

# Interpretation
print("\nInterpretation:")
print(f"We are {conf_level * 100}% confident that the true population mean height")
print(f"lies within the interval of {ci[0]:.2f} cm to {ci[1]:.2f} cm.")

# Another example with different data
weights = [60, 65, 70, 58, 72, 63, 68, 75, 61, 66, 70, 62]
mean_w, moe_w, ci_w = calculate_confidence_interval_mean(weights, confidence_level=0.90)

print(f"\n--- For Weights Dataset ({conf_level * 100}% Confidence) ---")
print(f"Sample Mean: {mean_w:.2f} kg")
print(f"Confidence Interval: ({ci_w[0]:.2f} kg, {ci_w[1]:.2f} kg)")
print(f"Interpretation: We are {0.90 * 100}% confident that the true population mean weight")
print(f"lies within the interval of {ci_w[0]:.2f} kg to {ci_w[1]:.2f} kg.")
```

```python
# 21. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

def calculate_and_interpret_ci_normal_data(population_mean, population_std_dev, sample_size, confidence_level=0.95):
    """
    Generates data from a normal distribution, calculates the confidence interval for its mean,
    and interprets the result.

    Args:
        population_mean (float): The true mean of the normal population from which data is drawn.
        population_std_dev (float): The true standard deviation of the normal population.
        sample_size (int): The number of data points to generate for the sample.
        confidence_level (float, optional): The desired confidence level (e.g., 0.95 for 95%).
                                            Defaults to 0.95.
    """
    if sample_size <= 1:
        raise ValueError("Sample size must be greater than 1.")
    if not (0 < confidence_level < 1):
        raise ValueError("Confidence level must be between 0 and 1.")

    # 1. Generate data from a normal distribution
    generated_data = np.random.normal(loc=population_mean, scale=population_std_dev, size=sample_size)

    # 2. Calculate sample statistics
    sample_mean = np.mean(generated_data)
    sample_std = np.std(generated_data, ddof=1) # Use ddof=1 for sample standard deviation

    # 3. Calculate the confidence interval
    # For CI of mean with unknown population std dev, use t-distribution
    t_critical = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)
    standard_error = sample_std / np.sqrt(sample_size)
    margin_of_error = t_critical * standard_error

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print(f"--- Confidence Interval for Mean (Generated Normal Data) ---")
    print(f"True Population Mean: {population_mean:.2f}")
    print(f"True Population Standard Deviation: {population_std_dev:.2f}")
    print(f"Sample Size (n): {sample_size}")
    print(f"Generated Sample Mean: {sample_mean:.2f}")
    print(f"Generated Sample Standard Deviation: {sample_std:.2f}")
    print(f"Confidence Level: {confidence_level * 100}%")
    print(f"Margin of Error: {margin_of_error:.2f}")
    print(f"Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

    # 4. Interpret the result
    print("\nInterpretation:")
    print(f"We are {confidence_level * 100}% confident that the true population mean")
    print(f"lies within the interval of {lower_bound:.2f} to {upper_bound:.2f}.")

    # Check if the true population mean falls within the calculated CI
    if lower_bound <= population_mean <= upper_bound:
        print(f"The true population mean ({population_mean:.2f}) is within the calculated confidence interval. (Expected)")
    else:
        print(f"The true population mean ({population_mean:.2f}) is NOT within the calculated confidence interval. (This happens with a probability of 1 - confidence_level)")

    # Optional: Visualize the sample and CI
    plt.figure(figsize=(10, 6))
    plt.hist(generated_data, bins=20, density=True, alpha=0.6, color='skyblue', edgecolor='black', label='Sample Data Histogram')
    plt.axvline(sample_mean, color='red', linestyle='--', label=f'Sample Mean: {sample_mean:.2f}')
    plt.axvline(lower_bound, color='green', linestyle=':', label='CI Lower Bound')
    plt.axvline(upper_bound, color='green', linestyle=':', label='CI Upper Bound')
    plt.axvspan(lower_bound, upper_bound, color='green', alpha=0.1, label=f'{confidence_level*100}% CI')
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.title(f"Generated Normal Data and {confidence_level*100}% Confidence Interval")
    plt.legend()
    plt.grid(axis='y', linestyle='--')
    plt.show()


# Example Usage:
calculate_and_interpret_ci_normal_data(
    population_mean=100,
    population_std_dev=15,
    sample_size=50,
    confidence_level=0.95
)

print("\n" + "="*50 + "\n")

calculate_and_interpret_ci_normal_data(
    population_mean=25,
    population_std_dev=5,
    sample_size=100,
    confidence_level=0.90
)
```

```python
# 23. Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameter for the Poisson distribution (average rate of events)
lambda_param = 3.5

# Define a range of values (k) for which to calculate the CDF
# Start from 0 up to a value where CDF is close to 1
max_k = int(poisson.ppf(0.999, lambda_param)) + 2 # Find value where CDF is almost 1
k_values = np.arange(0, max_k)

# Calculate the CDF values
cdf_values = poisson.cdf(k_values, lambda_param)

print(f"--- Cumulative Distribution Function (CDF) for Poisson Distribution (λ = {lambda_param}) ---")
print("-" * 70)
print(f"{'Number of Events (k)':<20} {'P(X <= k) (CDF)':<20} {'Interpretation':<30}")
print("-" * 70)

for k, cdf_val in zip(k_values, cdf_values):
    interpretation = f"The probability of observing {k} or fewer events is {cdf_val:.4f}."
    print(f"{k:<20} {cdf_val:<20.4f} {interpretation:<30}")

print("-" * 70)

# Visualize the CDF
plt.figure(figsize=(10, 6))
plt.step(k_values, cdf_values, where='post', color='darkcyan', marker='o', linestyle='-')
plt.xlabel("Number of Events (k)")
plt.ylabel("P(X <= k)")
plt.title(f"Cumulative Distribution Function (CDF) of Poisson Distribution (λ = {lambda_param})")
plt.xticks(k_values)
plt.yticks(np.linspace(0, 1, 11))
plt.grid(True, linestyle='--')
plt.show()

# Interpretation examples:
print("\nSpecific Interpretations:")
# Probability of exactly 2 events: P(X=2) = P(X<=2) - P(X<=1)
prob_exact_2 = poisson.pmf(2, lambda_param)
print(f"Probability of exactly 2 events: P(X=2) = {prob_exact_2:.4f}")

# Probability of at most 4 events: P(X <= 4)
prob_at_most_4 = poisson.cdf(4, lambda_param)
print(f"Probability of at most 4 events: P(X <= 4) = {prob_at_most_4:.4f}")

# Probability of more than 5 events: P(X > 5) = 1 - P(X <= 5)
prob_more_than_5 = 1 - poisson.cdf(5, lambda_param)
print(f"Probability of more than 5 events: P(X > 5) = {prob_more_than_5:.4f}")
```

```python
# 25. Write a Python program to compare the standard deviations of two datasets and visualize the difference.

import numpy as np
import matplotlib.pyplot as plt

def compare_and_visualize_std_devs(dataset1, dataset2, name1="Dataset 1", name2="Dataset 2"):
    """
    Calculates the standard deviations of two datasets and visualizes their distributions
    to show the difference in spread.

    Args:
        dataset1 (array-like): The first dataset.
        dataset2 (array-like): The second dataset.
        name1 (str, optional): Name for dataset 1. Defaults to "Dataset 1".
        name2 (str, optional): Name for dataset 2. Defaults to "Dataset 2".
    """
    data1 = np.array(dataset1)
    data2 = np.array(dataset2)

    std_dev1 = np.std(data1)
    std_dev2 = np.std(data2)

    mean1 = np.mean(data1)
    mean2 = np.mean(data2)

    print(f"--- Comparison of Standard Deviations ---")
    print(f"{name1} - Mean: {mean1:.2f}, Standard Deviation: {std_dev1:.2f}")
    print(f"{name2} - Mean: {mean2:.2f}, Standard Deviation: {std_dev2:.2f}")

    if std_dev1 > std_dev2:
        print(f"\nConclusion: {name1} has a larger spread (higher standard deviation) than {name2}.")
    elif std_dev2 > std_dev1:
        print(f"\nConclusion: {name2} has a larger spread (higher standard deviation) than {name1}.")
    else:
        print(f"\nConclusion: Both datasets have approximately the same standard deviation.")

    # Visualization
    plt.figure(figsize=(12, 6))

    # Histogram for Dataset 1
    plt.hist(data1, bins=20, density=True, alpha=0.6, color='blue', edgecolor='black', label=f'{name1} (Std Dev: {std_dev1:.2f})')
    # Histogram for Dataset 2
    plt.hist(data2, bins=20, density=True, alpha=0.6, color='red', edgecolor='black', label=f'{name2} (Std Dev: {std_dev2:.2f})')

    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.title("Comparison of Dataset Distributions (Differences in Spread)")
    plt.legend()
    plt.grid(axis='y', linestyle='--')
    plt.show()

# Example Usage:

# Dataset 1: Narrower spread
data_set_A = np.random.normal(loc=10, scale=2, size=500)

# Dataset 2: Wider spread
data_set_B = np.random.normal(loc=10, scale=5, size=500)

compare_and_visualize_std_devs(data_set_A, data_set_B, "Dataset A", "Dataset B")

print("\n" + "="*50 + "\n")

# Example 2: Different means but similar spread
data_set_C = np.random.normal(loc=50, scale=7, size=500)
data_set_D = np.random.normal(loc=60, scale=7.5, size=500)

compare_and_visualize_std_devs(data_set_C, data_set_D, "Scores C", "Scores D")
```

```python
# 26. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

import numpy as np

# Generate data from a normal distribution
mean = 100
std_dev = 15
num_samples = 500
normal_data = np.random.normal(loc=mean, scale=std_dev, size=num_samples)

# --- Calculate the Range ---
# Range = Maximum Value - Minimum Value
data_range = np.max(normal_data) - np.min(normal_data)

# --- Calculate the Interquartile Range (IQR) ---
# IQR = Q3 - Q1
# Q1: 25th percentile
# Q3: 75th percentile

q1 = np.percentile(normal_data, 25)
q3 = np.percentile(normal_data, 75)
iqr = q3 - q1

print(f"Dataset generated from Normal Distribution (Mean={mean}, Std Dev={std_dev}, Samples={num_samples})")
print(f"Minimum Value: {np.min(normal_data):.2f}")
print(f"Maximum Value: {np.max(normal_data):.2f}")
print(f"Calculated Range: {data_range:.2f}")
print(f"First Quartile (Q1 - 25th percentile): {q1:.2f}")
print(f"Third Quartile (Q3 - 75th percentile): {q3:.2f}")
print(f"Calculated Interquartile Range (IQR): {iqr:.2f}")

# Optional: Verify using scipy.stats.iqr for convenience (but manual calculation is shown above)
from scipy.stats import iqr as scipy_iqr
iqr_scipy = scipy_iqr(normal_data)
print(f"IQR (using scipy.stats.iqr): {iqr_scipy:.2f}")
```

```python
# 27. Implement Z-score normalization on a dataset and visualize its transformation.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def z_score_normalize_and_visualize(dataset, name="Original Data"):
    """
    Performs Z-score normalization on a dataset and visualizes the original
    and transformed distributions.

    Args:
        dataset (array-like): The input dataset to normalize.
        name (str, optional): A name for the dataset for plotting. Defaults to "Original Data".
    """
    original_data = np.array(dataset)

    # Calculate mean and standard deviation of the original data
    original_mean = np.mean(original_data)
    original_std = np.std(original_data)

    if original_std == 0:
        print("Standard deviation is zero. Cannot perform Z-score normalization.")
        return

    # Perform Z-score normalization
    # Z = (x - mean) / std_dev
    normalized_data = (original_data - original_mean) / original_std

    # Calculate mean and standard deviation of the normalized data (should be ~0 and ~1)
    normalized_mean = np.mean(normalized_data)
    normalized_std = np.std(normalized_data)

    print(f"--- Z-score Normalization Results for '{name}' ---")
    print(f"Original Data Mean: {original_mean:.2f}")
    print(f"Original Data Standard Deviation: {original_std:.2f}")
    print(f"Normalized Data Mean: {normalized_mean:.4f} (should be close to 0)")
    print(f"Normalized Data Standard Deviation: {normalized_std:.4f} (should be close to 1)")

    # Visualization
    plt.figure(figsize=(14, 6))

    # Plot original data histogram
    plt.subplot(1, 2, 1)
    plt.hist(original_data, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')
    plt.axvline(original_mean, color='red', linestyle='--', label=f'Mean: {original_mean:.2f}')
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.title(f"Original Distribution of {name}")
    plt.legend()
    plt.grid(axis='y', linestyle='--')

    # Plot normalized data histogram
    plt.subplot(1, 2, 2)
    plt.hist(normalized_data, bins=30, density=True, alpha=0.7, color='lightgreen', edgecolor='black')
    # Overlay standard normal PDF for comparison
    x_norm = np.linspace(min(normalized_data)-0.5, max(normalized_data)+0.5, 100)
    plt.plot(x_norm, norm.pdf(x_norm, loc=0, scale=1), color='purple', linestyle=':', linewidth=2, label='Standard Normal PDF')
    plt.axvline(normalized_mean, color='red', linestyle='--', label=f'Mean: {normalized_mean:.2f}')
    plt.xlabel("Z-score")
    plt.ylabel("Density")
    plt.title(f"Normalized Distribution (Z-scores)")
    plt.legend()
    plt.grid(axis='y', linestyle='--')

    plt.tight_layout()
    plt.show()

# Example Usage:

# Example 1: Data from a normal distribution
data1 = np.random.normal(loc=50, scale=10, size=1000)
z_score_normalize_and_visualize(data1, "Test Scores")

print("\n" + "="*50 + "\n")

# Example 2: Data from a non-normal distribution (e.g., exponential)
# Z-score normalization works for any distribution, but the shape will be preserved.
# Only the mean becomes 0 and std dev becomes 1.
data2 = np.random.exponential(scale=3, size=1000)
z_score_normalize_and_visualize(data2, "Waiting Times")
```

```python

# 28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(dataset, name="Dataset"):
    """
    Calculates the skewness and kurtosis of a given dataset.

    Args:
        dataset (array-like): The input dataset.
        name (str, optional): A name for the dataset for printing. Defaults to "Dataset".
    """
    data = np.array(dataset)

    # Calculate skewness
    # A value of 0 indicates perfect symmetry.
    # Positive skewness: long tail to the right (right-skewed).
    # Negative skewness: long tail to the left (left-skewed).
    calculated_skewness = skew(data)

    # Calculate kurtosis
    # For normal distribution, excess kurtosis is 0 (Fisher's kurtosis).
    # Positive kurtosis (leptokurtic): more outliers, heavier tails, sharper peak than normal.
    # Negative kurtosis (platykurtic): fewer outliers, lighter tails, flatter peak than normal.
    calculated_kurtosis = kurtosis(data) # By default, scipy.stats.kurtosis calculates Fisher's kurtosis (excess kurtosis)

    print(f"--- Skewness and Kurtosis for '{name}' ---")
    print(f"Number of data points: {len(data)}")
    print(f"Mean: {np.mean(data):.2f}")
    print(f"Standard Deviation: {np.std(data):.2f}")
    print(f"Skewness: {calculated_skewness:.4f}")
    print(f"Kurtosis (Fisher, Excess Kurtosis): {calculated_kurtosis:.4f}")

    # Interpretation
    print("\nInterpretation:")
    if abs(calculated_skewness) < 0.1: # Threshold for 'approximately symmetric'
        print("Skewness is close to 0, indicating the distribution is approximately symmetric.")
    elif calculated_skewness > 0:
        print(f"Skewness is positive, indicating the distribution is positively (right) skewed.")
    else: # calculated_skewness < 0
        print(f"Skewness is negative, indicating the distribution is negatively (left) skewed.")

    if abs(calculated_kurtosis) < 0.1: # Threshold for 'approximately mesokurtic'
        print("Kurtosis is close to 0 (excess kurtosis), indicating the distribution is approximately mesokurtic (similar peakedness to normal).")
    elif calculated_kurtosis > 0:
        print(f"Kurtosis is positive, indicating the distribution is leptokurtic (more peaked and heavier tails than normal).")
    else: # calculated_kurtosis < 0
        print(f"Kurtosis is negative, indicating the distribution is platykurtic (less peaked and lighter tails than normal).")

# Example Usage:

# 1. Dataset from a normal distribution (expected skewness ~0, kurtosis ~0)
normal_data = np.random.normal(loc=0, scale=1, size=1000)
calculate_skewness_kurtosis(normal_data, "Normal Distribution Data")

print("\n" + "="*50 + "\n")

# 2. Dataset from a right-skewed distribution (e.g., Exponential)
exponential_data = np.random.exponential(scale=1, size=1000)
calculate_skewness_kurtosis(exponential_data, "Exponential Distribution Data")

print("\n" + "="*50 + "\n")

# 3. Dataset from a distribution with high kurtosis (e.g., t-distribution with low df)
from scipy.stats import t
t_data = t.rvs(df=3, size=1000) # t-distribution with 3 degrees of freedom
calculate_skewness_kurtosis(t_data, "Student's t-Distribution (df=3) Data")
```
